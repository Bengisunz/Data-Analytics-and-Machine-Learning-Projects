```python
#Matplotlib is the whole package; matplotlib.pyplot is a module in matplotlib; and pylab is a module that gets installed alongside matplotlib.


x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()

plt.show()
```

```python
plt.hist(satis_df['tutar'])
plt.xlabel('tutar')
plt.show()
```

```python
for col in ('tutar','maliyet'):
    plt.hist(satis_df[col], normed=True, alpha=0.5)
plt.show()    
```

```python
ulke_bolge=pd.DataFrame({'count' : satis_df.groupby( [ "ulkeadi", "merchbolge"] ).size()}).reset_index()
ulke_bolge
```

```python
labels = ulke_bolge['ulkeadi']
sizes = ulke_bolge["count"]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral','blue','black','pink', 'g','r']

patches, texts = plt.pie(sizes, colors=colors,shadow=True, startangle=90)

plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = ulke_bolge['ulkeadi']
sizes = ulke_bolge["count"]
explode = (0,0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

# ---------------------------------SEABORN---------------------------------

Matplotlib'e bağlı, ayrıntılı figürler çizmeyi sağlayan istatistiksel veri görselleştirme kütüphanesi

*Seaborn harnesses the power of matplotlib to create beautiful charts in a few lines of code. The key difference is Seaborn’s default styles and color palettes, which are designed to be more aesthetically pleasing and modern. Since Seaborn is built on top of matplotlib, you’ll need to know matplotlib to tweak Seaborn’s defaults*


```python
# density plot with shade
sns.kdeplot(satis_df['maliyet'], shade=True)
#sns.plt.show()
```

```python
sns.countplot(y='merchbolge',data=ulke_bolge,palette="Set2")
plt.xlabel('Frekans')
plt.ylabel('Bolgeler')
plt.title('Bolgelere gore Dagılım')
plt.show()
```

```python
sns.countplot(x='merchbolge',data=ulke_bolge)
plt.xlabel('Bolgeler')
plt.ylabel('Frekans')
plt.title('Bolgelere gore Dagılım')
plt.show()
```

```python
sns.countplot(x='merchbolge',hue='ulkeadi',data=satis_df)
plt.xlabel('Bolgeler')
plt.ylabel('Frekans')
plt.title('Bolgelere gore Dagılım')

plt.show()
```

```python
ax = sns.scatterplot(x="satis_adet", y="tutar", hue="merchbolge",style="merchbolge",  data=satis_df)
#https://seaborn.pydata.org/generated/seaborn.scatterplot.html
```

```python
satis_corr=satis_df[["satis_adet","tutar","maliyet"]]
sns.pairplot(satis_corr, kind="scatter")
plt.show()
```

```python
# Calculate correlation between each pair of variable
corr_matrix=satis_corr.corr()
 
# plot it
sns.heatmap(corr_matrix, cmap='PuOr')
#sns.plt.show()
```

```python
#https://seaborn.pydata.org/introduction.html
sns.set()
sns.relplot(x="satis_adet", y="tutar", col="merchbolge", size="maliyet",data=satis_df);
```

```python
ax = sns.lineplot(x="YilAyGun", y="Satis", data=magaza_df)
plt.xticks(rotation=45)
```
